# Data Preparation

## Encoding Fitur Kategori

Melakukan encoding menggunakan Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
df_car['brand'] = labelencoder.fit_transform(df_car['brand'])
df_car['name'] = labelencoder.fit_transform(df_car['name'])
df_car['bodyType'] = labelencoder.fit_transform(df_car['bodyType'])
df_car['color'] = labelencoder.fit_transform(df_car['color'])
df_car['fuelType'] = labelencoder.fit_transform(df_car['fuelType'])
df_car['transmission'] = labelencoder.fit_transform(df_car['transmission'])
df_car['vehicleConfiguration'] = labelencoder.fit_transform(df_car['vehicleConfiguration'])
df_car['engineName'] = labelencoder.fit_transform(df_car['engineName'])
df_car['engineDisplacement'] = labelencoder.fit_transform(df_car['engineDisplacement'])
df_car['location'] = labelencoder.fit_transform(df_car['location'])

df_car
```

## Train-Test-Split

Membagi data train dan test menjadi 90 : 10

```python
from sklearn.model_selection import train_test_split
 
X = df_car.drop(["price"],axis =1)
y = df_car["price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 154)
```

```python
print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in train dataset: {len(X_train)}')
print(f'Total # of sample in test dataset: {len(X_test)}')
```

## Standarisasi

Melakukan standarisasi menggunakan Standard Scaler

```python
from sklearn.preprocessing import StandardScaler

numerical_features = ['year', 'mileage', 'power']
scaler = StandardScaler()
scaler.fit(X_train[numerical_features])
X_train[numerical_features] = scaler.transform(X_train.loc[:, numerical_features])
X_train[numerical_features].head()
```

# Modeling

## Modeling dengan KNN

```python
# import library yang digunakan
from sklearn.neighbors import KNeighborsRegressor

# membuat model prediksi
knn = KNeighborsRegressor(n_neighbors=13)
knn.fit(X_train, y_train)
```

Kita menggunakan k = 13 tetangga dan metric Euclidean untuk mengukur jarak antara titik

## Modeling dengan Random Forest

```python
# Impor library yang dibutuhkan
from sklearn.ensemble import RandomForestRegressor
 
# buat model prediksi
RF = RandomForestRegressor(n_estimators=100, max_depth=32, random_state=55, n_jobs=-1)
RF.fit(X_train, y_train)              
```

## Modeling dengan Boosting Algoritm

```python
# import library yang digunakan
from sklearn.ensemble import AdaBoostRegressor

# membuat model prediksi
boosting = AdaBoostRegressor(n_estimators=100, learning_rate=0.8, random_state=55)                             
boosting.fit(X_train, y_train)
```

# Evaluasi Model

```python
# Lakukan scaling terhadap fitur numerik pada X_test
X_test.loc[:, numerical_features] = scaler.transform(X_test[numerical_features])
```

```python
# import library yang digunakan
from sklearn.metrics import mean_squared_error

# Buat variabel mse yang isinya adalah dataframe nilai mse data train dan test pada masing-masing algoritma
rmse = pd.DataFrame(columns=['train', 'test'], index=['KNN','RF','Boosting'])
 
# Buat dictionary untuk setiap algoritma yang digunakan
model_dict = {'KNN': knn, 'RF': RF, 'Boosting': boosting}
 
# Hitung Mean Squared Error masing-masing algoritma pada data train dan test
for name, model in model_dict.items():
    rmse.loc[name, 'train'] = mean_squared_error(y_true=y_train, y_pred=model.predict(X_train), squared=False)
    rmse.loc[name, 'test'] = mean_squared_error(y_true=y_test, y_pred=model.predict(X_test), squared=False)
 
# Panggil mse
rmse
```

Plot metrik tersebut dengan bar chart

```python
fig, ax = plt.subplots()
rmse.sort_values(by='test', ascending=False).plot(kind='barh', ax=ax, zorder=3)
ax.grid(zorder=0)
```

Dari gambar di atas, terlihat bahwa, model Random Forest memberikan nilai error yang paling kecil. Sedangkan model dengan algoritma Boosting memiliki eror yang paling besar. Model Random Forest yang akan kita pilih sebagai model terbaik untuk melakukan prediksi harga mobil.

```python
prediksi = X_test.iloc[:15].copy()
pred_dict = {'y_true':y_test[:15]}
for name, model in model_dict.items():
    pred_dict['prediksi_'+name] = model.predict(prediksi).round(1)
 
pd.DataFrame(pred_dict)
```
